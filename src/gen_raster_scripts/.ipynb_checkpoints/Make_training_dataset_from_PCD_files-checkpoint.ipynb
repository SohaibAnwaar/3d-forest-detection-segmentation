# Plot Predicted Points

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_boxes_on_image(image, boxes, labels=None, scores=None):
    
    # Create a figure and axis
    fig, ax = plt.subplots(1)

    # Display the image
    ax.imshow(image)

    # Plot the bounding boxes
    for i, box in enumerate(boxes):
        
        try:
#             [ymin xmin ymax xmax]
            y1, x1, y2, x2 = box
            width = x2 - x1
            height = y2 - y1

            rect = patches.Rectangle((x1, y1), width, height, linewidth=1, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

            if labels is not None:
                label = labels[i]

                ax.text(x1, y1, label, color='white', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))
        except: pass

    plt.show()



# labels = ['object1', 'object2']
# img, boxes, classes, scores

bb_coord = np.transpose(np.vstack((boxes[:, 0] * grid_size[1], boxes[:, 1] * grid_size[0], boxes[:, 2] * grid_size[1],boxes[:, 3] * grid_size[0])))
plot_boxes_on_image(img, bb_coord, classes, scores)
```

# Predict PCD File

```python


grid_size = [600, 600, 1000]
grid_res = 0.1
output_dir = '../../data/'
# config_dict = {'raster_layers': ['vertical_density', 'canopy_height'], 'support_window': [11, 7, 1],
#                'normalisation': 'rescale+histeq','doHisteq':[False,True],'res': 0.1, 'gridSize': }

config_dict = {"doHisteq": [False, True], "raster_layers": ["vertical_density", "canopy_height"], "gridSize": grid_size, "res": grid_res, "support_window": [11, 7, 1], "normalisation": "rescale+histeq"}
rasterMaker = treeDetector.RasterDetector(**config_dict)


# PCD file
path = "../../data/150_2-160_2_slam.pcd"
save_bbox = "../../data/150_2-160_2_slam.txt"
pcd_file = read_pcd(path)
xyz_data = pcd_file['points'][['x','y','z']].values
offset = [0,0,0]





box_store = np.zeros((1, 4))
counter = 0
windowSize = [50,50]
stepSize = 40

xyz_data_gr = ground_removal.removeGround(xyz_data,offset,thresh=2.0,proc_path=output_dir)
ground_pts = ground_removal.load_ground_surface(os.path.join(output_dir,'_ground_surface.ply'))

xyz_clr_data = xyz_data
slices = detection_tools.sliding_window_3d(xyz_clr_data, stepSize=stepSize,windowSize=windowSize)


```

# Predicting Slices

```python

detector_addr = '../../models/models/detection/hvp/'
all_bbox = []

while  True:
    
    try:
        x, y, window = next(slices)
        if window is not None:
            # Rasterizing the Image slice
            image, center =rasterMaker._rasterise(window, ground_pts=ground_pts)
            image = np.uint8(image*255)
            plt.imshow(image)
            plt.show()

            # Predicting Image
            img, boxes, classes, scores = detectObjects(image, addr_weights=os.path.join(detector_addr,'yolov3.weights'),
                                                                  addr_confg=os.path.join(detector_addr,'yolov3.cfg'),MIN_CONFIDENCE=0.5)

            # Mapping 2d boxes to 3d
            bb_coord = detection_tools.boundingBox_to_3dcoords(boxes_=boxes, gridSize_=grid_size[0:2], gridRes_=grid_res,
                                                                          windowSize_=windowSize, pcdCenter_=center)



            overlap_thresh = 5
            idx = detection_tools.find_unique_boxes2(bb_coord, overlap_thresh)
            bb_coord = bb_coord[idx, :]
            all_bbox.extend(bb_coord)
            print("bbox", bb_coord)
            
            
            bb_coord_percent_to_pixel = np.transpose(np.vstack((boxes[:, 0] * grid_size[1], boxes[:, 1] * grid_size[0], boxes[:, 2] * grid_size[1],boxes[:, 3] * grid_size[0])))
            plot_boxes_on_image(img, bb_coord_percent_to_pixel, classes, scores)

    except IndexError as index_error:
        print("Error Boxes", boxes)
    
    except StopIteration: break
        


with open(save_bbox, "w") as f:
    np.savetxt(f, np.asarray(all_bbox))

```
